```python
import sympy as sy
```

```python
x,y = sy.symbols('x,y', real = True)
```

```python
f = sy.exp(-x**2/2) * sy.cosh(y)
f
```

```python
import matplotlib
sy.plotting.plot3d(f,(x,-5,5),(y,-5,5))
```

```python
fx = sy.diff(f,x)
fx
```

```python
fy = sy.diff(f,y)
fy
```

```python
sy.solve([fx,fy],[x,y])
```

```python
sy.plot(sy.cosh(x),(x,-4,4))
```

```python
sy.plot(sy.sinh(x),(x,-2,2))
```

```python
H = sy.hessian(f,(x,y))
H
```

```python
H.subs({x:0, y:0})
```

```python
gamma, r, D, alpha = sy.symbols('gamma, r, D, alpha')
```

```python
W = sy.S(1)/2 * gamma * D * r**2 * (alpha-sy.sin(alpha))
W
```

```python
gamma_m = 8.75
r_m = 10.0
D_m = 0.1
alpha_m = sy.pi/3
W_est = W.subs({gamma: gamma_m,
                r: r_m,
                D: D_m,
                alpha: alpha_m})
```

```python
W_est
```

```python
W_est.evalf()
```

```python
errors = []
for i in range(4):
    x = [gamma, r, D, alpha][i]
    y = [gamma_m, r_m, D_m, alpha_m][i]
    errors.append(sy.diff(W,x).subs(x,y).evalf())
errors
```
