## Plotting multivariate functions

### a) Manual 3D plots with Matplotlib

In the following, we are going to generate a 2D plot of the function $$f(x,y) = x^2 + y^2 + 2x$$
over the domain $$[-2,2]\times [-2,2].$$

We start by defining all pairs of $(x,y)$-values. The corresponding command is `np.meshgrid()`


```python
# Generate x and y values using np.meshgrid()
x = np.linspace(-2,2,100)
y = np.linspace(-2,2,100)
X, Y = np.meshgrid(x,y)
```

```python
type(X)
```

```python
X
```

```python
# Create a figure and a 3D graphic
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection='3d')
Z = X**2 * Y**2+ 2*X
ax.plot_surface(X,Y,Z)
ax.set_title('Surface plot of f(x,y) = x^2 + y^2 + 2x')
```

It is possible and useful to make surface plots interactive. In a "normal" Jupyter Notebook it is enough to use the magic `%matplotlib widget` to activate interactivity.

```python
# Generate the same graphic as above, this time as an interactive graphic
fig = plt.figure(figsize=(12,10))
ax = plt.axes(projection='3d')
Z = X**2 * Y**2+ 2*X
ax.plot_surface(X,Y,Z)
ax.set_title('Surface plot of f(x,y) = x^2 + y^2 + 2x')
```

```python
fig = plt.figure()
cp = plt.contour(X,Y,Z)
ax = plt.gca()
fig.colorbar(cp)
ax.set_title('Contour plot of f(x,y) = x^2 + y^2 + 2x')
```

```python
fig = plt.figure()
cp = plt.contourf(X,Y,Z)
ax = plt.gca()
fig.colorbar(cp)
ax.set_title('Filled contour plot of f(x,y) = x^2 + y^2 + 2x')
```

### b) Automatic 3D plot with SymPy

```python
import sympy as sy
```

```python
x,y = sy.symbols('x,y')
f = x**2 + y**2 + 2*x
sy.plotting.plot3d(f,(x,-2,2),(y,-2,2))
```
