## Plotting a 2D curve

```python
import sympy as sy
```

```python
t = sy.symbols('t')
```

```python
sy.plotting.plot_parametric(sy.exp(t)*sy.cos(t), sy.exp(t)*sy.sin(t),(t,0,5))
```

## Plotting a 3D curve

```python
import sympy as sy
t = sy.symbols('t')
```

```python
sy.plotting.plot3d_parametric_line(sy.cos(t),sy.sin(t),t,(t,0,4*sy.pi))
```

## Exercise 8.1

```python
import sympy as sy
t = sy.symbols('t')
```

```python
v = (sy.diff(sy.exp(t)*sy.cos(t),t)**2 +
            sy.diff(sy.exp(t)*sy.sin(t),t)**2)**(sy.S(1)/2)
v.simplify()
```

```python
sy.integrate(v,(t,-sy.oo,1)).simplify()
```

```python
b = sy.symbols('b')
sy.integrate(v,(t,b,1)).simplify()
```

```python
V = sy.integrate(v,t).simplify()
V
```

```python
sy.diff(V,t)
```
