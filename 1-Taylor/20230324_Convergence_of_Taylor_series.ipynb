## 1) Python exercise
Compute different Taylor polynomials of the natural logarithm at $x_0 = 1$. Plot these and the logarithm itself on the interval $(0.1, 2.2)$ (respective $(-0.1,2.2)$ for $T_n$).

From these experiments, derive an hypothesis where the Taylor series of the natural logarithm at $x_0 = 1$ is a convergent series.

```python
import sympy as sy
x = sy.symbols('x')
```

```python
Tn = sy.series(sy.ln(x),x,1,15).removeO()
sy.plot((sy.log(x),(x,.1,2.2)),(Tn,(x,-.1,2.2)),title = 'T15')
```

```python
Tn = sy.series(sy.ln(x),x,1,20).removeO()
sy.plot((sy.log(x),(x,.1,2.2)),(Tn,(x,-.1,2.2)),title = 'T20')
```

```python
Tn = sy.series(sy.ln(x),x,1,25).removeO()
sy.plot((sy.log(x),(x,.1,2.2)),(Tn,(x,-.1,2.2)),title = 'T55')
```

```python
Tn = sy.series(sy.ln(x),x,1,100).removeO()
sy.plot((sy.log(x),(x,.1,1.9)),(Tn,(x,-.01,1.9)),title = 'T100')
```
