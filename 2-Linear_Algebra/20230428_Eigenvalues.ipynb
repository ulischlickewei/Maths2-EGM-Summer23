```python
A = sy.Matrix([[0,-1],[1,0]])
A
```

```python
l = sy.symbols('l')
```

```python
# Manually compute the characteristic polynomial
(A - l*sy.eye(2)).det()
```

```python
# Use Python to compute the characteristic polynomial
A.charpoly(l).as_expr()
```

```python
# Determine the eigenvalues of A
sy.solve(A.charpoly(l).as_expr(),l)
```

Why does $A = \begin{pmatrix} 0 & -1\\ 1 & 0 \end{pmatrix}$ does not have real eigenvalues?

Because multiplying an arbitrary real vector $v$ with this matrix give as a result the vector $v$ rotated by 90 degrees in counterclockwise direction. After rotation by 90 degrees this cannot be collinear with $v$.

```python
# Determine eigenspace corresponding to the eigenvalue i
(A - sy.I*sy.eye(2)).rref()[0]
```

```python
# Determine eigenspace corresponding to the eigenvalue -i
(A + sy.I*sy.eye(2)).rref()[0]
```

```python
# Use SymPy internal functions for computing eigenvalues and eigenvectors
A.eigenvals()
```
