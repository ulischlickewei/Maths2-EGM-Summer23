```python
import sympy as sy
```

## Identity matrix

```python
M = sy.Matrix([[3,-4],[7,13]])
M
```

```python
I = sy.eye(2)
I
```

```python
I*M
```

```python
M*I
```

```python
v = sy.Matrix([1,17])
v
```

```python
I*v
```
