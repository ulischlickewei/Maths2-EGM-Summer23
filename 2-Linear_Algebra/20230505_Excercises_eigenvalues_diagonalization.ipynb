```python
import sympy as sy
a,b = sy.symbols('a,b')
```

```python
A = sy.Matrix([[a,b],[b,a]])
A
```

```python
A.eigenvects()
```

Compute a diagonalization of $A$, i.e. provide $P$, $SD$ and $P^{-1}$ such that $$A = P D P^{-1}$$

```python
P = sy.Matrix([[-1,1],[1,1]])
P
```

```python
D = sy.Matrix([[a-b, 0],[0,a+b]])
D
```

```python
P*D*P.inv()
```

```python
A.diagonalize()
```
